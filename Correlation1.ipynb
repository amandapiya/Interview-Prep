```python
import csv
```

```python
log_occurances = {}
bigM = 0
prevMin = 0
with open('1067fabb-6c5e-48b7-a4ed-be7663308407.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=' ', quotechar='|')
    for row in spamreader:
        id, time = row[0].split(",")
        if time and time[3] != 'e':
            minute = int(time[-5] + time[-4])
            if minute < prevMin:
                max_key = int(max(log_occurances, key=log_occurances.get))
                bigM = max(bigM, max_key)
                log_occurances = {}
                print(bigM)
            else:
                if id in log_occurances:
                    log_occurances[id] += 1
                else:
                    log_occurances[id] = 0
            prevMin = minute
            
```
